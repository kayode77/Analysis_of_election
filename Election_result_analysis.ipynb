# Analysis of Nigerian presidential election in 2011,2015 and 2019. With indept analysis of the 2015 presidential election in the states.

```python
import pandas as pd
A=pd.read_csv('Nigerian presidential election, 2019',index_col=0 , sep=',')
A.head()
```

```python
A.info()
```

```python
A.describe()
```

```python
A.iat[0,2],A.iat[1,2],A.iat[2,2],A.iat[3,2]
```

```python
Others=  A['%'].sum()-(A.iat[0,2]+A.iat[1,2]+A.iat[2,2]+A.iat[3,2])
Others
```

```python
A.iat[0,2],A.iat[1,2],A.iat[2,2],A.iat[3,2]
Others=  A['%'].sum()-(A.iat[0,2]+A.iat[1,2]+A.iat[2,2]+A.iat[3,2])
import matplotlib.pyplot as plt 
fig=plt.figure(figsize=(10, 7))
ax=fig.add_axes([0,0,1,1])
Aggregatescore = [A.iat[0,2],A.iat[1,2],A.iat[2,2],A.iat[3,2],Others]
Candidates = ['Muhammadu Buhari', 'Atiku Abubakar', 'Felix Nicolas', 'Obadiah Mailafia','Others']
colors = ['green', 'gold', 'blue', 'purple','red']
explode = (0.0, 0.0, 0.2, 0.0, 0.0) 
plt.pie(Aggregatescore, labels=Candidates, colors=colors,explode=explode, autopct='%1.2f%%',shadow=True) 
centre_circle = plt.Circle((0,0),0.80,color='black', fc='white',linewidth=1.00)
fig = plt.gcf() 
fig.gca().add_artist(centre_circle)
plt.axis('equal')
```

```python
# A plot for 2019
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
g = sns.PairGrid(A)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
plt.show()
```

# Analysis of 2015 general election in Nigeria

```python
import pandas as pd
A3=pd.read_csv('Nigerian presidential election 2, 2015', index_col=0, sep=',')
A3.head()
```

```python
A3.columns=[['Muhammadu Buhari','Goodluck Jonathan','Adebayo Ayeni','Ganiyu Galadima','Sam Eke','Rufus Salau','Mani Ahmad','Allagoa Chinedu',
  'Martin Onovo','Tunde Anifowose-Kelani','Chekwas Okorie','Comfort Sonaiya','Godson Okoye','Ambrose Albert Owuru']]
```

```python
A3
```

```python
A3=A3.T
A3
```

```python
A3.index.names = ['Candidates']
A3.columns.names = ['']
A3
```

```python
AC=A3.loc[['Muhammadu Buhari','Goodluck Jonathan','Adebayo Ayeni']]
AC
```

# Virtualization of three leading candidates in the election of 2015

```python
AC1=AC.iloc[0][:-1]
AC1
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.set_xlabel('Votes Obtained')
ax.set_title('Votes Obtained by Muhammadu Buhari in 2015 election candidate of All Progressives Congress plot')
AC1.plot.barh(stacked=True,alpha=0.5,color='r')
```

```python
AC2=AC.iloc[1][:-1]
AC2
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.set_xlabel('Votes Obtained')
ax.set_title("Votes Obtained by Goodluck Jonathan in 2015 election candidate of People's Democratic Party plot")
AC2.plot.barh(stacked=True,alpha=0.5,color='g')
```

```python
AC3=AC.iloc[2][:-1]
AC3
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.set_xlabel('Votes Obtained')
ax.set_title("Votes Obtained by Adebayo Ayeni in 2015 election candidate of African Peoples Alliance Party plot")
AC3.plot.barh(stacked=True,alpha=0.5,color='b')
```

Therefore, Kano, Katsina and Kaduna are strongholds in the North for any aspiring candidate in winning any election in Nigeria.
